## Part2: Load data and perform Bioinformatics Analyses (pairwise sequence and blast)

[Description of code below, reading sequences from a fasta file.]
 A fasta file is a FASTA formatted text based format for representing sequences, either nucleotide or peptide, in my case nucleotide. It is called FASTA because of "fast-all" meaning that this type of file works with any coding alphabet. 
 
In this code below we are reading nucleotide sequences from a fasta file and comparing them to each other to find similarities between their sequences. We use the pairwise2 alignment tool to find these similarities. 

We are also using the fasta file sequences for perfom blast searches to compare the nucleotide sequences with a library of sequences and find local similarites between the sequences. 

```python
#open each file and create variable for each species

FastaB = list(SeqIO.parse("BelugaMT.txt","fasta"))
Almas = FastaB[0].seq

# length of sequence
print(len(Almas))
print(Almas)

FastaA= list(SeqIO.parse("ATSalmonMT.txt","fasta"))
Salmon = FastaA[0].seq
print(len(Salmon))
print(Salmon)

# perform pairwise alignment
# name the x and y variables to be aligned
X = Almas
Y = Salmon

#use globalxx to find best alignment between sequences
alignments = pairwise2.align.globalxx(X,Y)

for a in alignments:
    print(format_alignment(*a))
```

```python
# Blast search

# sturgeon blast
Fasta = list(SeqIO.parse("BelugaMT.txt", "fasta"))
Sturg = Fasta[0].seq
print(Sturg)

# New sequence was saved as a fasta file with the original sequence after doing a BLAST search. 
Fasta1 = list(SeqIO.parse("ATSalmonMT.txt", "fasta"))
BLAST_hit = Fasta1[0].seq
print(BLAST_hit)

# Setting the window size and which sequence is on the x and y axes of the dot plot
window = 3
seq_one = str(Sturg)
seq_two = str(BLAST_hit)

# Saving the data in the dot plot based on window size and comparing for all I and all j in the range of the lengths of the
# sequences, wheter or not there is an alignment.

data = [
    [
        (seq_one[i : i + window] != seq_two[j: j + window])
        for j in range(len(seq_one) - window)
    ]
    for i in range(len(seq_two) - window)
]

import pylab

pylab.gray()
pylab.imshow(data)
pylab.xlabel("%s (length %i bp)" % (Fasta[0].id, len(Fasta[0].seq)))
pylab.ylabel("%s (length %i bp)" % (Fasta1[0].id, len(Fasta1[0].seq)))
pylab.title("Dot plot using window size %i\n(allowing no mis-matches)" % window)
pylab.show()
```

# Create heatmap showing similarities

Below we will be utilizing matplotlib to visualize the similarities from the pairwise sequence alignment. There should be axes and a color bar. The role of heatmaps are to show the similarity of at leastr relative similarity between the scores from the dataset. Here it is used to compare the alignment scores between sequences. 

```python
#Similar code as from week 3 challenge problem
FastaB = list(SeqIO.parse("BelugaMT.txt","fasta"))
Almas = FastaB[0].seq

FastaA= list(SeqIO.parse("ATSalmonMT.txt","fasta"))
Salmon = FastaA[0].seq

size = len(FastaB)

#Initiate variable i and j for for loop
i = 0
j = 0

#Initiate an empy numpy matrix that is the length of the fasta file
alignments_score = np.empty(shape=(size,size))

#Nested for loop to do pairwise sequence alignment for each pair of kinases
for i in range(size):
    for j in range(size):
        # Define two sequences to be aligned
        X = FastaB[i].seq
        Y = FastaA[j].seq
        print(FastaB[i].id)
        print(FastaA[j].id)
        
        # Get a list of the global alignments between the two sequences 
        # No parameters. Identical characters have score of 1, else 0.
        # No gap penalties.
        alignments_score[i,j] = int(pairwise2.align.globalxx(X, Y, score_only=True))
# Check that the for loop correctly created the matrix
#print(alignments_score)
print(alignments_score)
#Create subplot
fig, ax = plt.subplots()

# Actually show the heatplot
plt.imshow(alignments_score)

# Instruct the heat map to show the axes
im = ax.imshow(alignments_score)

# Add a colorbar to the right of the heatmap
cbar = ax.figure.colorbar(im)

# Show the plot
plt.show()
```
